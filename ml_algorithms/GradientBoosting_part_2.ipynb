```python
from sklearn.datasets import make_regression
import pandas as pd
from sklearn.metrics import mean_absolute_error
```

```python
X, y = make_regression(n_samples=10, n_features=3)
```

```python
df = pd.DataFrame(X)
```

```python
df['y_true'] = y 
```

```python
df
```

```python
df['y_pred'] = df['y_true'].mean()
```

```python
df
```

```python
mean_absolute_error(df['y_true'], df['y_pred'])
```

```python
from sklearn.tree import DecisionTreeRegressor
lr = 0.1
n = 10
trees = []

for i in range(n):
    df['residual'] = df['y_true'] - df['y_pred']
    tree = DecisionTreeRegressor(max_depth=1)
    tree.fit(df[[0,1,2]], df['residual'])
    df['y_pred'] = df['y_pred'] + lr * tree.predict(df[[0,1,2]])
    trees.append(tree)
    print(mean_absolute_error(df['y_true'], df['y_pred']))
```

```python
df
```

```python
test = df[[0,1,2]].copy()
```

```python
test['y_pred'] = df['y_true'].mean()
```

```python
for tree in trees:
    test['y_pred'] = test['y_pred'] + lr * tree.predict(df[[0,1,2]])
```
